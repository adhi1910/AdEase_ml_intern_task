#### Logo, Company name, call to action from the image

```python
import numpy as np

img = cv2.imread('web_scrapping.jpg')

blank = np.zeros(img.shape[:2], dtype='uint8')

mask=cv2.rectangle(blank,(10,10),(100,100),255,-1)

masked = cv2.bitwise_and(img, img, mask=mask)
cv2.imshow('Output', masked)

img_crop=masked[10:100,10:100] 
cv2.imwrite('web_logo_1.jpg',img_crop)
cv2.waitKey(0)
```
